4- Calcule la media, la varianza, la asimetría, la curtosis, los cuantiles, la mediana y el rango intercuartílico
de A y B.

- Estadisticos para Datos de concentración mg/kg de la Columna A

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
from google.colab import drive
from scipy.stats import kurtosis, skew

# Montar Google Drive
drive.mount('/content/drive')

# Leer los datos del archivo .csv
ruta = "/content/drive/MyDrive/1_Maestria_en_recursos_Hidraulicos/Tercer_Semestre/Metodos_Estocasticos/Trabajos/T1_Estadistica_descriptiva/SMwr_Ex_01.csv"
datos = pd.read_csv(ruta)

# Extraer la columna 'A'
columna_A = datos['A'].values

# Calcular las estadísticas
media = np.mean(columna_A)
varianza = np.var(columna_A)
simetria = skew(columna_A)
curtosis = kurtosis(columna_A)
cuartiles = np.percentile(columna_A, [25, 50, 75])
mediana = np.median(columna_A)
rango_intercuartilico = cuartiles[2] - cuartiles[0]

# Imprimir las estadísticas
print(f"Media: {media} mg/Kg")
print(f"Varianza: {varianza} mg²/Kg²")
print(f"Simetría: {simetria}")
print(f"Curtosis: {curtosis}")
print(f"Cuartiles: {cuartiles} mg/Kg")
print(f"Mediana: {mediana} mg/Kg")
print(f"Rango intercuartilico: {rango_intercuartilico} mg/Kg")
```

5- Dibuje un diagrama de caja y bigotes de los valores de A y B. ¿Hay posibles valores atípicos?

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt

# Montar Google Drive
drive.mount('/content/drive')

# Leer los datos del archivo .csv
ruta = "/content/drive/MyDrive/1_Maestria_en_recursos_Hidraulicos/Tercer_Semestre/Metodos_Estocasticos/Trabajos/T1_Estadistica_descriptiva/SMwr_Ex_01.csv"
datos = pd.read_csv(ruta)

# Extraer las columnas 'A' y 'B'
columna_A = datos['A'].values
columna_B = datos['B '].values

# Crear los diagramas de cajas y bigotes
plt.figure(figsize=(8, 6))
plt.boxplot([columna_A, columna_B], vert=True, labels=["Grupo A", "Grupo B"])
plt.title('Diagramas de Cajas y Bigotes para las Columnas A y B')
plt.ylabel('Concentración mg/Kg')
plt.show()

# Función para detectar valores atípicos
def detectar_valores_atipicos(datos):
    Q1 = np.percentile(datos, 25)
    Q3 = np.percentile(datos, 75)
    IQR = Q3 - Q1
    valor_minimo = Q1 - 1.5 * IQR
    valor_maximo = Q3 + 1.5 * IQR
    valores_atipicos = datos[(datos < valor_minimo) | (datos > valor_maximo)]
    return valores_atipicos

# Detectar y mostrar valores atípicos en las columnas 'A' y 'B'
valores_atipicos_A = detectar_valores_atipicos(columna_A)
valores_atipicos_B = detectar_valores_atipicos(columna_B)

if valores_atipicos_A.size:
    print("Valores atípicos en el grupo de datos de concentración A (mg/Kg):")
    print([str(valor) + ' mg/Kg' for valor in valores_atipicos_A])
else:
    print("No se reportan valores atípicos en el grupo de datos de concentración A")

if valores_atipicos_B.size:
    print("Valores atípicos en el grupo de datos de concentración B (mg/Kg):")
    print([str(valor) + ' mg/Kg' for valor in valores_atipicos_B])
else:
    print("No se reportan valores atípicos en el grupo de datos de concentración B.")
```

6-Suponga que A es la concentración de algún contaminante en el suelo (mg/kg). Supongamos que las
muestras se han tomado aleatoriamente del sitio de interés. Si la concentración crítica es de 5 mg/kg y
el sitio tiene 8000 m2. ¿Qué área aproximada del sitio debe ser limpiada?

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt

# Montar Google Drive
drive.mount('/content/drive')

# Leer los datos del archivo .csv
ruta = "/content/drive/MyDrive/1_Maestria_en_recursos_Hidraulicos/Tercer_Semestre/Metodos_Estocasticos/Trabajos/T1_Estadistica_descriptiva/SMwr_Ex_01.csv"
datos = pd.read_csv(ruta)

# Extraer los valores de la columna 'A'
columna_A = datos['A'].values


# Calcular el porcentaje de datos en la columna A que son mayores o iguales a 5mg/Kg
porcentaje_de_concentracion_critica = np.sum(columna_A >= 5) / len(columna_A) * 100

# Calcular el área que necesita ser limpiada basándose en el porcentaje calculado y el área total del sitio
area_total_sitio = 8000  # en metros cuadrados
area_a_limpiar = area_total_sitio * (porcentaje_de_concentracion_critica / 100)

print(f"El porcentaje de datos en el grupo de datos A que son mayores o iguales a 5mg/Kg es: {porcentaje_de_concentracion_critica}%")
print(f"El área aproximada que debe ser limpiada en el sitio es: {area_a_limpiar} m²")
```
