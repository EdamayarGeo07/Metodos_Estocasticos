<font size="3" >5. Dibuje un diagrama de caja y bigotes de los valores de A y B. ¿Hay posibles valores atípicos?

```python
#Estilo de grafica
plt.style.use('classic')

# Crear los diagramas de cajas y bigotes
plt.figure(figsize=(8, 6))
plt.boxplot([columna_A, columna_B], vert=True, labels=["Grupo A", "Grupo B"])
plt.title('Diagramas de Cajas y Bigotes para las Columnas A y B')
plt.ylabel('Concentración mg/Kg')
plt.show()

# Función para detectar valores atípicos
def detectar_valores_atipicos(datos):
    Q1 = np.percentile(datos, 25)
    Q3 = np.percentile(datos, 75)
    IQR = Q3 - Q1
    valor_minimo = Q1 - 1.5 * IQR
    valor_maximo = Q3 + 1.5 * IQR
    valores_atipicos = datos[(datos < valor_minimo) | (datos > valor_maximo)]
    return valores_atipicos

# Detectar y mostrar valores atípicos en las columnas 'A' y 'B'
valores_atipicos_A = detectar_valores_atipicos(columna_A)
valores_atipicos_B = detectar_valores_atipicos(columna_B)

if valores_atipicos_A.size:
    print("Valores atípicos en el grupo de datos de concentración A (mg/Kg):")
    print([str(valor) + ' mg/Kg' for valor in valores_atipicos_A])
else:
    print("No se reportan valores atípicos en el grupo de datos de concentración A")

if valores_atipicos_B.size:
    print("Valores atípicos en el grupo de datos de concentración B (mg/Kg):")
    print([str(valor) + ' mg/Kg' for valor in valores_atipicos_B])
else:
    print("No se reportan valores atípicos en el grupo de datos de concentración B.")
```
